```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
breast_cancer_data=sklearn.datasets.load_breast_cancer()
print(breast_cancer_data)
```

```python
df=pd.DataFrame(breast_cancer_data.data,columns= breast_cancer_data.feature_names)
```

```python
df.head()
```

```python
df['class']=breast_cancer_data.target
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df['class'].value_counts()
```

```python
X=df.drop(columns='class', axis=1)
Y=df['class']
```

```python
x_train,x_test,y_train,y_test= train_test_split(X,Y, test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_std=scaler.fit_transform(x_train)
x_test_std=scaler.fit_transform(x_test)
```

```python
print(x_train_std.shape)
print(x_test_std.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
tf.random.set_seed(4)
```

```python
model= tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(30,)),
    tf.keras.layers.Dense(25,activation='relu'),
    tf.keras.layers.Dense(10,activation='relu'),
    tf.keras.layers.Dense(2,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy']) 
```

```python
pred=model.fit(x_train_std,y_train, validation_split=0.15, epochs=100)
```

```python
plt.plot(pred.history['accuracy'])
plt.plot(pred.history['val_accuracy'])

plt.title('Accuracy plot')
plt.ylabel('accuracy')
plt.xlabel('epochs')

plt.legend(['training data', 'validation data'])
```

```python
plt.plot(pred.history['loss'])
plt.plot(pred.history['val_loss'])

plt.title('Loss plot')
plt.ylabel('loss')
plt.xlabel('epochs')

plt.legend(['training data', 'validation data'])
```

```python
loss, accuracy= model.evaluate(x_test_std,y_test)
```

```python
y_pred=model.predict(x_test_std)
print(y_pred)
```
